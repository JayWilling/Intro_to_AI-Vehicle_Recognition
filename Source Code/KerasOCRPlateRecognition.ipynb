```python
import os
import matplotlib.pyplot as plt

import keras_ocr as keras_ocr
```

```python
# keras-ocr will automatically download pretrained
# weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
# Get a set of three example images
images = [
    keras_ocr.tools.read(url) for url in [
        #'detected/0_ncjz03.jpg',
        #'detected/0_xyn5n2.jpg',
        'detected/0_0ok5hw.jpg',
        'detected/0_u0pubs.jpg',
        'detected/1_0ok5hw.jpg'
    ]
]
```

```python
images = [
    keras_ocr.tools.read('detected/' + image) for image in os.listdir('detected/')
]
```

```python
# Each list of predictions in prediction_groups is a list of
# (word, box) tuples.
prediction_groups = pipeline.recognize(images)
```

```python
# Plot the predictions
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
